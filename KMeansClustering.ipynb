## Importing Libraries 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
```

## Loading dataset

```python
iris = load_iris()
X=iris.data
```

## Selecting Optimum value of K using Elbow method:

```python
wcss=[]
for i in range(1,11):
    k=KMeans(n_clusters=i,random_state=0)
    k.fit(X)
    wcss.append(k.inertia_)
```

```python
plt.figure(figsize=(11,5),facecolor='skyblue')
plt.plot(range(1,11),wcss)
plt.xlabel('No of Cluster')
plt.ylabel("WCSS")
plt.title("Elbow Method")
```

<h3>it looks like elbow point is at x=3, 
    So, k=3
</h3>

```python
km=KMeans(n_clusters=3)
```

## fitting the model

```python
km.fit(X)
```

## Predicting 

```python
y_pred= km.predict(X)
y_pred
```

## Visualizing the clusters

```python
plt.figure(figsize=(11,5),facecolor='lime',edgecolor='black')
plt.scatter(X[y_pred==0,0],X[y_pred==0,1],color='red',marker="*")
plt.scatter(X[y_pred==1,0],X[y_pred==1,1],color='orange')
plt.scatter(X[y_pred==2,0],X[y_pred==2,1],color='blue',marker="+")

plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='green')
plt.legend(['Iris-setosa','Iris-versicolour','Iris-virginica','Centroid'])
```
